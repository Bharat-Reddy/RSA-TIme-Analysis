```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
key = np.array([1024, 2*1024, 3*1024, 4*1024, 5*1024])
time = np.array([10**2, 10**3, 10**4, 10**5, 10**6, 10**7, 10**8]) #microseconds
gen_time = np.array([145276, 271143, 461271, 1716317, 3289729])
dec_time = np.array([7363, 7695, 10937, 20991, 36848])
enc_time = np.array([1911, 1045, 853, 1685, 1593])
```

```python
plt.yscale("log")
plt.tick_params(axis='both', which='major', labelsize=14)
plt.rcParams["figure.figsize"] = (10,10)

plt.plot(key, gen_time, 'r-', label="Generation", marker='o')
plt.plot(key, dec_time,'g-', label="Decryption", marker='o')
plt.plot(key, enc_time, 'b-', label="Encryption", marker='o')

#create legend
plt.legend(loc="upper left")

#labeling
plt.xlabel('Key size', fontsize=20)
plt.ylabel('Time (microseconds)', fontsize=20)
plt.title("Time for RSA key operations as a function of key size ", fontsize=20);
```
